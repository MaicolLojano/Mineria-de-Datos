## Nota: En este caso existen algunos valores de texto en edad y Fecha inscripcio, tenemos que crear los valores a numerico para generar nuestro reporte.

```python
# Convertir la columna 'Edad' a valores numéricos, los valores no convertibles se transforman en NaN
df['Edad'] = pd.to_numeric(df['Edad'], errors='coerce')
```

```python
# Rellenar los valores NaN en la columna 'Edad' con 0
df['Edad'].fillna(0, inplace=True)
```

```python
# Verificamos los cambios
print(df['Edad'].head())
```

```python
# Convertir la columna 'Fecha_Inscripcion' a fechas, los valores no convertibles se transforman en NaT
df['Fecha_Inscripcion'] = pd.to_datetime(df['Fecha_Inscripcion'], errors='coerce')
```

```python
# Verificamos los cambios
print(df['Fecha_Inscripcion'].head())
```

```python
advert_report=sv.analyze([df,"REPORTE DE DATOS_ESTUDIANTES"])
```

```python
advert_report.show_html()
```

## MANEJO DE VALORES FALTANTES

# 1. Eliminar filas con valores faltantes

```python
# Manejo de datos faltantes
# Contar el número de filas con valores faltantes
num_filas_con_nulos = df.isnull().sum(axis=1)
# Mostrar el resultado
print("Número de filas con valores faltantes:")
print(num_filas_con_nulos)
```

```python
df_sin_nulos = df.dropna()
print("\nDataFrame sin filas con valores faltantes:")
print(df_sin_nulos.head())
```

```python
df_sin_nulos.describe()
```

```python
#Guadar la data en un archivo excel
df_sin_nulos.to_excel("Datos_sin_nulos.xlsx")
```

# 2. Rellenar valores faltantes con un valor específico (por ejemplo, la media)

```python
# Identificar columnas con valores faltantes
missing_values = df.isnull().sum()
print("Valores faltantes por columna:")
print(missing_values)
```

## Observacion: En este caso no existen valores faltantes, por lo cual no hace falta rellenar con un valor en específico.

# 3. Interpolación para estimar valores faltantes

```python
df_interpolado = df.interpolate()
print("\nDataFrame con valores faltantes interpolados:")
print(df_interpolado.head())
```

```python
#Guadar la data en un archivo excel
df_interpolado.to_excel("Datos_interpolados.xlsx")
```

# 4. Eliminar columnas con valores faltantes


```python
df_sin_nulos_columnas = df.dropna(axis=1)
print("\nDataFrame sin columnas con valores faltantes:")
print(df_sin_nulos_columnas.head())
```

```python
#Guadar la data en un archivo excel
df_sin_nulos_columnas.to_excel("Datos_sin columnas.xlsx")
```
